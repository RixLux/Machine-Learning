```python
# Create dataframe to summarise missing values
def missing value_pct_df(data):
    percent_missing = data.isnull().mean()
    missing value _df = pd.DataFrame(percent_missing).reset_index()
    missing value_df = missingvalue_df.rename(columns = {‘index':'column_name’, Q: ‘percent_missing'})
    #muLltiply by 100 and round to 4 decimals places
    missingvalue_df[‘percent_missing']=missing value_df[‘percent_missing’].apply(lambda x: round(x*100,2))
    missing value_df = missing value_df.sort_values(by = [‘percent_missing'], ascending = False)
    return missing value _df
```

```python
missing value df = missing value _pct_df(user_data)
missing value_df
```

```python
#thershold above which to drop feature
threshold = 80
cols_to_drop = list(missing value _df[missing value_df[‘percent_missing'] > threshold][‘column_name’])
print(‘Number of features to drop :', missing _value_df[missing value_df['percent_missing'] > threshold].shape[Q])
print(f'Features with missing values greather than {threshold}%:', cols_to_drop)
```

```python
user_data_cleaned = user_data.drop(cols_to_drop, axis=1)
```

```python
missing value df = missing valuepct_df(user_data_cleaned)
missing column = list(missing value _df[missing value df['percent_missing'] > 0] ['column_name'])
print('Number of features with missing values:', len(missing_column))
```
